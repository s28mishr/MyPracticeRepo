```
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```
np.random.seed(101)
tf.set_random_seed(101)
```

```
x = np.linspace(0,50,50)
y = np.linspace(0,50,50)
```

```
x += np.random.uniform(-4,4,50)
y += np.random.uniform(-4,4,50)
```

```
plt.scatter(x, y) 
plt.xlabel('x') 
plt.xlabel('y') 
plt.title("Training Data") 
plt.show() 
```

```
X = tf.placeholder(float)
Y = tf.placeholder(float)
```

```
W = tf.Variable(np.random.randn(), name = "W") 
b = tf.Variable(np.random.randn(), name = "b") 
```

```
learning_rate = 0.01
train_epoch = 1000
```

```
y_hat = tf.add(tf.multiply(X,W), b)
```

```
cost = tf.reduce_sum(tf.pow((y_hat-Y),2))/ (2*(len(x)))
```

```
optimizer =  tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
```

```
init = tf.global_variables_initializer()
```

```
with tf.Session() as sess:
  sess.run(init)
  for epoch in range(train_epoch):
    for (a,b) in zip(x,y):
      sess.run(optimizer, feed_dict = {X: a, Y: b})
      
      if (epoch+1) % 50 == 0:
        c = sess.run(cost, feed_dict = {X : x, Y : y}) 
        print("Epoch=", epoch+1, " Cost=", c, " Weight=", c, " Bias=",b)
  training_cost = sess.run(cost, feed_dict ={X: x, Y: y}) 
  print(training_cost)
  weight = sess.run(W)
  print(weight)
  constant = sess.run(b)
  print(constant)
```

```
predictions = weight * x + constant
print("Training Cost=", training_cost, " weight=", weight, " bias=", constant, "y-hat", predictons)
```
